## In this Jupyter Notebook, I've utilized the Pandas library to analyze a dataset containing reviews of various wines. The dataset includes information such as the country of origin, description, designation, points, price, province, region, taster details, title, variety, and winery.

## Throughout this notebook, I've performed a variety of activities such as reading the dataset, writing to it, indexing, selecting specific data, assigning values, and leveraging functions like map() and apply() to manipulate and transform the data.

```python
import pandas as pd
```

```python
file_path = '../data/wine.csv'
wine_df = pd.read_csv(file_path)
```

```python
wine_df.head(5)
```

```python
wine_df.shape
```

### let us only take 100 rows for simplicity

```python
wine_sub = wine_df.loc[:100]
wine_sub.tail()
```

### Getting information from 2nd row

```python
wine_sub.iloc[2]
```

### selecting the country column just from the first, second, and Fourth row

```python
wine_sub.loc[[0,1,3],'country']
```

### selecting country, price, province, and taster_name from 5th to 10th row

```python
wine_sub.loc[5:10,['country','price','province','taster_name']]
```

### selecting wine only from Italy

```python
wine_sub.loc[wine_sub.country == 'Italy']
```

### selecting wines from Italy and having points greater than 87

```python
wine_sub.loc[(wine_sub.country == 'Italy') & (wine_sub.points > 87)]
```

### selecting wines either from France or having exactly 88 points

```python
wine_sub.loc[(wine_sub.country == 'France') | (wine_sub.points == 88)]
```

### selecting wines only from Italy or France

```python
wine_sub.loc[wine_sub.country.isin(['Italy','France'])]
```

### listing number of wines each country have in wine_sub dataframe

```python
wine_sub.country.value_counts()
```

### let's take mean from points column and subtract it from every row

```python
points_mean = wine_sub.points.mean()
```

```python
wine_sub.points.map(lambda p: p - points_mean)
```

### Now let's apply the above in wine_sub dataframe

```python
def mean_point(row):
    row.points = row.points - points_mean
    return row

wine_sub.apply(mean_point, axis = 'columns').head()
```

### lets see the wine having the highest points to price ratio which is the best bargain

```python
bargain_id = (wine_sub.points/wine_sub.price).idxmax()
wine_sub.loc[bargain_id]
```

### listing the price of the cheapest wine from every country

```python
wine_sub.groupby('country').price.min()
```

### listing the best wine produced in every  country from each province

```python
wine_sub.groupby(['country','province']).apply(lambda df: df.loc[df.points.idxmax()])
```
