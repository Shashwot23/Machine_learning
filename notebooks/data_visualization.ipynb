```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## This csv is about fifa ranking of 5 different countries from 1993-2018

```python
file_path = "../data/fifa.csv"
fifa_data = pd.read_csv(file_path, index_col="Date", parse_dates=True)
```

```python
fifa_data.head()
```

```python
#set the width and height of the figure
plt.figure(figsize=(14,6))

#add title
plt.title("FIFA rankings of these 5 country")

#line chart
sns.lineplot(data = fifa_data)
```

```python
plt.figure(figsize=(14,8))
plt.title("Fifa rankings of Brazil and Germany")
sns.lineplot(data=fifa_data["BRA"], label="Brazil")
sns.lineplot(data=fifa_data['GER'], label="Germany")
plt.xlabel("Date")
             
```

### IGN is website that reviews games and the followings are their data

```python
ign_filepath = "../data/ign_scores.csv"
ign_data = pd.read_csv(ign_filepath, index_col="Platform")
ign_data
```

# Bar Graph

### lets take Racing Genre and see its average ratings in different platform

```python
plt.figure(figsize=(18,10))
sns.barplot(x=ign_data.index, y=ign_data['Racing'])
```

### from the bar graph we can see racing games are not good in Wii platform meanwhile they are greate for Xbox One

# Heatmap

```python
plt.figure(figsize=(10,10))
sns.heatmap(data=ign_data, annot=True)
plt.xlabel("Genre")
plt.title("Average Game Score")
```

###  Simulation games for Playstation 4 receive the highest average ratings (9.2). Shooting and Fighting games for Game Boy Color receive the lowest average rankings (4.5).
